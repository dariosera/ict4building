```python
import pandas as pd
import dec
from epw import epw
```

```python
meteo=pd.read_csv('IPIEMONT220.csv',index_col='Time',parse_dates=[0,])
meteo.head()
```

```python
meteo=meteo.resample('H').mean().interpolate()
meteo.head()
```

```python
meteo.columns
```

```python
latitude = 45.0619883  # latitude  N
longitude = 7.6602814  # longitude E
elevation = 250  # meters above the sea
```

```python
rad=pd.DataFrame(meteo['SolarRadiationWatts/m^2'].values,index=meteo.index,columns=['GHI'])
```

```python
rad=dec.main(rad,latitude,longitude,elevation)
```

```python
rad
```

```python
data=epw()
data.read('ITA_Torino.160590_IWEC.epw')
```

```python
head=data.headers # this is a dictionary of the header information
print(head) 
```

```python
df=data.dataframe.copy()  # this is pandas dataframe
```

```python
df.columns
```

```python
meteo.columns
```

```python
df['Dry Bulb Temperature']=meteo.TemperatureC.values
```

```python
df['Dew Point Temperature']=meteo.DewpointC.values
```

```python
df['Atmospheric Station Pressure'].unique()
```

```python
meteo.PressurehPa.unique()
```

In EPW the Pressure is expressed in Pa in out meteo data the Pressure is in hPa (Hectopascal) so 1Pa = 100*hPa

```python
df['Atmospheric Station Pressure']=100*meteo.PressurehPa.values
```

```python
df['Relative Humidity']=meteo.Humidity.values
```

```python
df['Global Horizontal Radiation']=rad.GHI.values
```

```python
df['Direct Normal Radiation']=rad.DNI.values
```

```python
df['Diffuse Horizontal Radiation']=rad.DHI.values
```

```python
df['Wind Direction']=meteo.WindDirectionDegrees.values
```

```python
df['Wind Speed']=meteo.WindSpeedKMH.values
```
